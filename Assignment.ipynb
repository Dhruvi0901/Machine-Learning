```python
## total null values in whole data set
df.isnull().sum().sum()
```

# **FILL NULL VALUES**

```python
## fill with 0
df1 = df.fillna(value = 0)
df1
```

```python
df1.isnull().sum().sum()
```

```python
## fill the value with previous value
df2 = df.fillna(method = 'pad')
df2
```

```python
## fill the null value with the mean of the column

df3 = df.fillna(value = df['fixed acidity'].mean())
df3
```

# **Dropna**

```python
df
```

```python
df4 = df.dropna()
df4
```

```python
## if all null then remove the row 
df5 = df.dropna(how='all')
df5
```

```python
## if any null remove that row
df6 = df.dropna(how = 'any')
df6
```

# **Replace**

```python
## replace the null value with any number
import numpy as np
df7 = df.replace(to_replace=np.nan,value = 67354)
df7
```

```python
## replace any value with any number
df8 = df.replace(to_replace=8.5, value = 9.5)
df8
```

# **Interpolate**

```python
## fill the null values using interpolate(automatically fill the values linearly )
df['sulphates'] = df['sulphates'].interpolate(method = 'linear')
df
```
