```python
#plotting histogram
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
df = pd.read_csv('/content/winequalityN.csv - winequalityN.csv.csv')
df["quality"].plot(kind = 'hist')
plt.show()
```

```python
#plotting histogram
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
df = pd.read_csv('/content/winequalityN.csv - winequalityN.csv.csv')
df['sulphates'].plot(kind = 'hist')
plt.show()
```

```python
#scatter plot
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/winequalityN.csv - winequalityN.csv.csv')
df.plot(kind = 'scatter', x = 'density', y = 'sulphates')
plt.show()
```

```python
#box plot
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(10)
data = np.random.normal(100, 20, 200)
fig = plt.figure(figsize =(10, 7))
plt.boxplot(data)
plt.show()
```

```python
np.random.seed(1234)
df = pd.DataFrame(np.random.randn(10, 4),
                  columns=['Col1', 'Col2', 'Col3', 'Col4'])
boxplot = df.boxplot(column=['Col1', 'Col2', 'Col3'])
```

```python
df = pd.DataFrame(np.random.randn(10, 2),
                  columns=['Col1', 'Col2'])
df['Y'] = pd.Series(['A', 'A', 'A','A'
                     'B', 'B', 'B','B'])
boxplot = df.boxplot(by='Y')
```
